```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time, re
import requests
import ast
import json
```

```python
# get url for all courses
driver = webdriver.Chrome()
driver.implicitly_wait(3)
driver.get(url='https://www.edx.org/course/?language=English&course=all')
for i in range(1,1000):
    time.sleep(1)
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
```

```python
soup = BeautifulSoup(driver.page_source,"html.parser")
```

```python
urllist = [ a['href'] for a in soup.find_all('a',class_="course-link") ]
```

```python
edXlist = []
count = 0
for course in urllist:
    print(count,course)
    idict = {}
    url  = requests.get(course)
    data = url.text
    csoup = BeautifulSoup(data,"html.parser")
    
    course_name = csoup.find("meta", property="og:title")["content"]
    description = csoup.find("meta", property="og:description")
    if description:
        description = description["content"]
    else:
        description = None
    image_url = csoup.find("meta", property="og:image")
    if image_url:
        image_url = image_url["content"]
    else:
        image_url = None
    course_url = csoup.find("meta", property="og:url")["content"]
    
    semantic = re.search('({"@context":).*',csoup.text).group(0)
    semantic = ast.literal_eval(semantic)
    subject = semantic['@graph'][1]['subjectOfStudy']
    provenance = semantic['@graph'][1]['provider']['name']
    
    idict["id"] = "edx"+format(count, '05d')
    idict["course_name"] = course_name
    idict["course_url"] = course_url
    idict["subject"] = subject
    idict["description"] = description
    idict["image_url"] = image_url
    idict["provenance"] = provenance
    
    edXlist.append(idict)
    count += 1
```

```python
with open('edX.json', 'w') as f:
    json.dump(edXlist, f, indent=4)
```
