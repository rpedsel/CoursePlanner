```python
import json
```

```python
with open('Catalogue.json') as f:
    data = f.read()
data = json.loads(data)
```

```python
cdict = {}
for program in data:
    for course in program['courses']:
         if course['course_id'] not in cdict.keys():
            cdict[course['course_id']] = course
```

```python
len(cdict)
```

```python
clist = []
for c in cdict:
    clist.append(cdict[c])
```

```python
clist
```

```python
with open('Catalogue_course.json', 'w') as outfile:
    json.dump(clist, outfile, indent=4)
```
