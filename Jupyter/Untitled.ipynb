```python
import numpy as np
table['price'] = table['price'].replace('Unknown', np.nan)
```

```python
table['price'].describe()
```

```python
table['price'] = pd.to_numeric(table['price'])
```

```python
table['price'].describe()
```

```python
table['Summary_Recalls'].describe()
```

```python
table['price_diff_per'] = (table['price']-table['Price_Dealers_Retail'])/table['Price_Dealers_Retail']
```

```python
table['price_diff_per'].describe()
```

```python
table['price_score'] = ((table['price_diff_per'].apply(lambda x: 1 if x >= 1 else x)) - table['price_diff_per'].min()) / (1 - table['price_diff_per'].min())
```

```python
table['price_score'] = 1 - table['price_score'] # reverse scale
```

```python
table[['price_diff_per', 'price_score']]
```

```python
table['mileage_year'] = table['Mileage'] / (2017 - table['year'].apply(lambda x: 2017 if x >= 2017 else x) + 1) 
# maybe year 2018 model
```

```python
table['mileage_year_score'] = 1 - (table['mileage_year'].apply(lambda x: 20000if x>=20000 else x) / 20000)
```

```python
table['mileage_year_score'].describe()
```

```python
table['year'].describe()
```

```python
df2 = pd.read_csv("truecar_url.csv")
df1 = pd.read_csv("edmunds_url.csv")
urls = pd.concat([df1, df2])
```

```python
table = table.join(urls.set_index('vin'), on='vin')
```

```python
table = table.drop_duplicates(subset=['vin'])
```

```python
table['recall_score'] = 1 - (table['Summary_Recalls']/(table['Summary_Recalls'].max() - table['Summary_Recalls'].min()))
```

```python
table['rating_score'] = (table['BODY & INTERIOR QUALITY - DESIGN'] + table['BODY & INTERIOR QUALITY - MECHANICAL'] + table['FEATURES & ACCESSORIES QUALITY - DESIGN'] + table['FEATURES & ACCESSORIES QUALITY - MECHANICAL'] + table['OVERALL QUALITY - DESIGN'] + table['OVERALL QUALITY - MECHANICAL'] + table['POWERTRAIN QUALITY - DESIGN'] + table['POWERTRAIN QUALITY - MECHANICAL']) / 40
```

```python
table['rating_score'].describe()
```

```python
table['score'] = table['rating_score'].apply(lambda x: 0 if x is np.nan else x) * 20 + \
table['recall_score'].apply(lambda x: 0 if x is np.nan else x) * 10 + \
table['mileage_year_score'].apply(lambda x: 0 if x is np.nan else x) * 30 + \
table['price_score'].apply(lambda x: 0 if x is np.nan else x) * 40
```
